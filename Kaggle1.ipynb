```python
for i in df['CarName']:
    a = i.split()
    df['CarName'].replace({i : a[0]}, inplace=True)

df
```

```python
df['CarName'].unique()
```

```python
df['CarName'].replace({'maxda' : 'mazda' ,
                       'vokswagen' : 'volkswagen' ,
                       'vw' : 'volkswagen' ,
                       'Nissan' : 'nissan' ,
                       'porcshce' : 'porsche' ,
                       'toyouta' : 'toyota'} ,
                       inplace=True)
```

```python
df['CarName'].unique()
```

```python
brand_count = df['CarName'].value_counts()
brand_count
```

```python
# Plotting the bar chart
plt.figure(figsize=(80,25))

plt.bar(brand_count.keys() , brand_count.values)

# Adding labels and title
plt.xlabel('Brand')
plt.ylabel('Count')
plt.xticks(fontsize=35)
plt.yticks(fontsize=40)

# Display the plot
plt.show()
```
